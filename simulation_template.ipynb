#### Simulation

```python
def simulate():
    var1 = np.random.poisson(lam=5,size=1)
    var2 = np.random.uniform(low=0.0,high=1.0,size=1)
    var3 = np.random.binomial(n=10,p=0.3,size=1)
    return var1+var2*var3
```

```python
def simulation(time):
    simulation_values = []
    for i in range(time):
        simulation_values.append(simulate())
    return simulation_values
```

```python
sim_arr = simulation(time=100)
plt.figure(figsize=(4,3))
plt.plot(sim_arr)
```
