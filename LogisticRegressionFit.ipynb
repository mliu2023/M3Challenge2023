```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

x = np.array(range(1,27))

y = np.array([3629.76966, 4493.87434, 5541.34286, 6799.56312, 8294.24472, 10046.1489, 12066.8124, 14353.5993, 16884.8761, 19616.6027, 22481.9211, 25395.0795, 28260.0095, 30982.2523, 33481.3923, 35700.625, 37611.0453, 39210.2268, 40516.579, 41561.88806, 42384.1526, 43022.1446, 43511.7948, 43884.4174, 44166.1333, 44378.0599])


def f(x, a, b, c, d):
    return a/(b + c*np.exp(-x*d))

(a_, b_, c_, d_), _ = opt.curve_fit(f, x, y)

y_fit = f(x, a_, b_, c_, d_)
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(x, y, 'o')
ax.plot(x, y_fit, '-')
```
