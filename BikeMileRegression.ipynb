```python
import numpy as np
import sklearn.linear_model
import pandas as pd
```

```python
import matplotlib.pyplot as plt
def LinearRegression(x, y):
  model = sklearn.linear_model.LinearRegression()
  model.fit(x,y)
  print(model.coef_, model.intercept_)
  plt.plot(x, y)
  plt.plot(x, model.coef_[0]*x + model.intercept_)
  plt.xlabel('Years after 2023')
  plt.ylabel('Total Bike Miles')
```

```python
df = pd.read_csv('TransportationMiles.csv')
df
```

```python
columns = [1, 3]
for column in columns:
  col = df[['Year', df.columns[column]]]
  col = col.dropna()
  x = np.array(col.iloc[:,0])
  x = np.expand_dims(x-2023, axis=1)
  y = np.array(col.iloc[:,1])
  #print(x)
  #print(y)
  LinearRegression(x,y)
```
